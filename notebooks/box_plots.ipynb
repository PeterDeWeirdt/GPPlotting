```python
import gpplotting
import seaborn as sns
```

```python
#Setting Style
sns.set_style('ticks')

#Adding additional settings for changing color and width of lines in boxplots
boxprops = {'edgecolor': 'black','linewidth':2}
lineprops = {'color': 'black','linewidth':2}
boxplot_kwargs = dict({'boxprops': boxprops, 'medianprops': lineprops,  
                   'whiskerprops': lineprops, 'capprops': lineprops,
                   'width': 0.75})
flierprops = dict(marker='o', markersize=5)
```

```python
tips = gpplotting.sns.load_dataset("tips")
tips.head()
```

```python
ax = sns.boxplot(x="size", y="total_bill", data=tips,palette="Set2")
```

```python
ax = sns.boxplot(x="size", y="total_bill", data=tips,palette="Paired")
```

```python
def stylized_boxplot(palette):
    ax = sns.boxplot(x="size", y="total_bill", data=tips,palette=palette,**boxplot_kwargs,flierprops=flierprops,
                    whis=[10, 90])
    ax.set_xlabel(ax.get_xlabel(),fontsize=20,color='black')
    ax.set_ylabel(ax.get_ylabel(),fontsize=20,color='black')

#     ax.tick_params(length=5, width=2,labelsize=20,color='black')
    ax.tick_params(direction='out', length=5, width=2, colors='black')
    for _,s in ax.spines.items():
        s.set_linewidth(2)
        s.set_color('black')

    ax.set_xticklabels(ax.get_xticklabels(),color='black',fontsize=20)
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_color('black') #set the color
        tick.label1.set_size(20) #set the size    

    for i,artist in enumerate(ax.artists):
        # Set the linecolor on the artist to the facecolor, and set the facecolor to None
        col = artist.get_facecolor()

        # Each box has 6 associated Line2D objects (to make the whiskers, fliers, etc.)
        # Loop over them here, and use the same colour as above
        for j in range(i*6,i*6+6):
            line = ax.lines[j]
    #         line.set_color(col)
            line.set_mfc(col)
            line.set_mec(col)
```

```python
stylized_boxplot('Set2')
```

```python
stylized_boxplot('Paired')
```
