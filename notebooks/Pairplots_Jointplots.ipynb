# Pairplots

Pairplots can be used when you want to plot data that share x and/or y axes. The x-axis will be shared across all plots in a row and the y-axis will be shared across all plots in a column. The diagonal axis will consist of plots showing the univariate distribution of all data for the variables in each column. The command and all its associated specifications are listed below, and can also be found at https://seaborn.pydata.org/generated/seaborn.pairplot.html.

seaborn.pairplot(data, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=True, plot_kws=None, diag_kws=None, grid_kws=None, size=None)

```python
plot = sns.pairplot(df, hue= "species", palette= "Paired")
```

The four variables sepal_length, sepal_width, petal_length and petal_width will be used in combination as the x and y axes. Species type will be distinguished by color by indicating hue = "species".

```python
plot = sns.pairplot(df, hue= "species", palette= "Set2")
```

# Jointplots

Jointplots can be used to include two types of plots (bivariate and univariate) in the same graph. The following info and more can also be found at https://seaborn.pydata.org/generated/seaborn.jointplot.html.

seaborn.jointplot(x, y, data=None, kind='scatter', stat_func=None, color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)

where kind can be: scatter | reg | resid | kde | hex (scatter is the default)

```python
plot2 = sns.jointplot('sepal_width', 'petal_length', data = df, kind = 'kde', space = 0, color = "#a6cee3")
```

There are two ways to set the color of your plot. Either you can indicate the exact color you want (as I did here using the first color in the Paired color scheme), or you can set the overall palette for seaborn which will automatically fill in any plot generated.

```python
#sns.set(palette= 'Paired')
#sns.set(palette= 'Set2')
```

```python
#using the first color in the Set2 color scheme
plot2 = sns.jointplot('sepal_width', 'petal_length', data = df, kind = 'kde', space = 0, color = "#66c2a5")
```
