```python
import gpplotting as gpp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

We can use the scatterplot function in seaborn to generate scatter plots with each category colored using a color from Set2 palette. 

```python
fig, ax = plt.subplots()
ax = sns.scatterplot(x='total_bill', y='tip', data=data, hue='time', palette='Set2')
t = sns.despine()
```

The different categories can also be plotted using different shapes. 

```python
ax = sns.scatterplot(x='total_bill', y='tip', data=data, hue='time', style='time', palette='Set2')
t = sns.despine()
```

```python
# Using Paired color palette
ax = sns.scatterplot(x='total_bill', y='tip', data=data, hue='time', style='time', palette='Paired')
t = sns.despine()
```

Another variable in the dataset can be used to make the size of the dot variable.

```python
# Changing the size of the dot uniformly
ax = sns.scatterplot(x='total_bill', y='tip', data=data, hue='time', palette='Set2',s=100)
t = sns.despine()
```

```python
# Changing the size of the dot based on a variable
ax = sns.scatterplot(x='total_bill', y='tip', data=data, size='size', palette='Set2', hue='time')
plt.legend(bbox_to_anchor=(1, 1)) #Placing the legend outside the plotting area
t = sns.despine()
```

Generating density plots using the function in GPPlotting

```python
gpp.gen_density_plot(data, 'total_bill', 'tip','RdBu')
```

```python
# Creating custom diverging palette
cmap = sns.diverging_palette(h_neg=221.61,h_pos=16.75, as_cmap=True)
gpp.gen_density_plot(data, 'total_bill', 'tip',cmap)
```

```python
# Creating custom diverging palette
cmap = sns.diverging_palette(h_neg=221.61,h_pos=16.75, as_cmap=True, center='dark')
gpp.gen_density_plot(data, 'total_bill', 'tip',cmap)
```

```python
# Using viridis, brighter center showing bulk of the population
gpp.gen_density_plot(data, 'total_bill', 'tip','viridis')
```
